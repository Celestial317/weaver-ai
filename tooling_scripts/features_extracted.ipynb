```python
from fashion_clip.fashion_clip import FashionCLIP
from PIL import Image
import torch

# Load FashionCLIP
device = "cuda" if torch.cuda.is_available() else "cpu"
fclip = FashionCLIP("fashion-clip")

# Thresholds
SIM_THRESHOLD = 0.20

# Your attribute options
CLOTHING_TYPES = [
    "Shirt", "T-shirt", "Sweater", "Jacket", "Blazer",
    "Hoodie", "Blouse"
]

GENDER_OPTIONS = [
    "Male", "Female", "Unisex", "Child"
]

OCCASION_OPTIONS = [
    "Casual", "Formal", "Party Wear"
]

SLEEVE_TYPES = [
    "Short Sleeve", "Long Sleeve", "Sleeveless"
]

NECKLINE_TYPES = [
    "Crew Neck", "V-Neck", "Collar", "High Neck", "Hooded", "N/A"
]

CLOSURE_TYPES = [
    "Zipper", "Button", "None", "Tie", "Hook and Eye", "Velcro", "N/A"
]

COLOR_OPTIONS = [
    "brown", "green", "orange", "yellow", "peach", "pink", "teal", "beige", "navy", "blue", "purple", "gray", "red", "black", "burgundy", "coral", "turquoise", "aqua", "gold", "mint"
]
PATTERN_TYPES = [
    "Solid", "Striped", "Floral", "Checked", "Polka Dot",
    "Graphic Print", "Animal Print", "Plaid", "Geometric", "N/A"
]

# Example pattern descriptions for additional detail
PATTERN_DESCRIPTIONS = {
    "Floral": ["Tiny floral motifs", "Large floral print"],
    "Striped": ["Thin horizontal stripes", "Thick vertical stripes"],
    "Checked": ["Small check pattern", "Large plaid squares"],
    "Polka Dot": ["Small polka dots", "Large polka dots"],
    "Animal Print": ["Leopard print", "Zebra stripes"],
    "Graphic Print": ["Abstract graphics", "Text slogans"],
    "Geometric": ["Triangle patterns", "Diamond shapes"],
    "Plaid": ["Classic tartan plaid"]
}

# Helper function for single-label attributes
def predict_single_attribute(image_feats, options, threshold=SIM_THRESHOLD):
    text_features_np = fclip.encode_text(options, batch_size=8)[0]
    text_features = torch.tensor(text_features_np).to(device)
    text_features = text_features / text_features.norm(dim=-1, keepdim=True)
    sims = (image_feats @ text_features.T).squeeze().tolist()

    if isinstance(sims, float):
        sims = [sims]

    best_idx = sims.index(max(sims))
    best_score = sims[best_idx]

    if best_score < threshold:
        return "N/A", best_score
    else:
        return options[best_idx], best_score


# Helper function for multi-label attributes
def predict_multi_labels(image_feats, options, threshold=SIM_THRESHOLD):
    text_features_np = fclip.encode_text(options, batch_size=8)[0]
    text_features = torch.tensor(text_features_np).to(device)
    text_features = text_features / text_features.norm(dim=-1, keepdim=True)



    sims = (image_feats @ text_features.T).squeeze().tolist()

    labels = []
    for label, score in zip(options, sims):
        if score >= threshold:
            labels.append(label)

    if not labels:
        return ["N/A"]
    else:
        return labels


# Main function
def analyze_garment(image_path):
    image = Image.open(image_path)
    image_feats_np = fclip.encode_images([image], batch_size=1)[0]
    image_feats = torch.tensor(image_feats_np).to(device)
    image_feats = image_feats / image_feats.norm(dim=-1, keepdim=True)


    # Predict attributes
    clothing_type, _ = predict_single_attribute(image_feats, CLOTHING_TYPES)
    gender_suitability = predict_multi_labels(image_feats, GENDER_OPTIONS)
    occasion_suitability = predict_multi_labels(image_feats, OCCASION_OPTIONS)
    sleeve_type, _ = predict_single_attribute(image_feats, SLEEVE_TYPES)
    neckline, _ = predict_single_attribute(image_feats, NECKLINE_TYPES)
    closure_type, _ = predict_single_attribute(image_feats, CLOSURE_TYPES)
    dominant_color, color_score = predict_single_attribute(image_feats, COLOR_OPTIONS)

    # Optionally try to get a secondary color
    secondary_color = "N/A"
    if dominant_color != "N/A":
        reduced_colors = [c for c in COLOR_OPTIONS if c != dominant_color]
        secondary_color, secondary_score = predict_single_attribute(image_feats, reduced_colors)
        if secondary_score < SIM_THRESHOLD:
            secondary_color = "N/A"

    pattern_type, pattern_score = predict_single_attribute(image_feats, PATTERN_TYPES)

    # Determine pattern description
    if pattern_type in PATTERN_DESCRIPTIONS:
        descriptions = PATTERN_DESCRIPTIONS[pattern_type]
        pattern_description, _ = predict_single_attribute(image_feats, descriptions, threshold=0.10)
    else:
        pattern_description = "N/A"

    garment_json = {
        "clothing_type": clothing_type,
        "gender_suitability": gender_suitability,
        "occasion_suitability": occasion_suitability,
        "sleeve_type": sleeve_type,
        "neckline": neckline,
        "closure_type": closure_type,
        "dominant_color": dominant_color,
        "secondary_color": secondary_color,
        "pattern_type": pattern_type,
        "pattern_description": pattern_description,
        "remarks": ""
    }

    return garment_json


# Example usage
if __name__ == "__main__":
    json_output = analyze_garment("/content/viton/clothes_tryon_dataset/test/cloth/00006_00.jpg")
    import json
    print(json.dumps(json_output, indent=4))
```

```python
import os
from PIL import Image
import torch
import numpy as np
import json
from transformers import CLIPProcessor, CLIPModel

# === Model Setup ===
device = "cuda" if torch.cuda.is_available() else "cpu"
model = CLIPModel.from_pretrained("patrickjohncyh/fashion-clip").to(device)
processor = CLIPProcessor.from_pretrained("patrickjohncyh/fashion-clip")

# === Attributes ===
CLOTHING_TYPES = [
    "Shirt", "T-shirt", "Sweater", "Jacket", "Blazer",
    "Hoodie", "Blouse"
]
GENDER_OPTIONS = [
    "Male", "Female", "Unisex", "Child"
]
OCCASION_OPTIONS = [
    "Casual", "Formal", "Party Wear"
]
SLEEVE_TYPES = [
    "Short Sleeve", "Long Sleeve", "Sleeveless"
]
NECKLINE_TYPES = [
    "Crew Neck", "V-Neck", "Collar", "High Neck", "Hooded", "N/A"
]
CLOSURE_TYPES = [
    "Zipper", "Button", "None", "Tie", "Hook and Eye", "Velcro", "N/A"
]
COLOR_OPTIONS = [
    "brown", "green", "orange", "yellow", "peach", "pink", "teal",
    "beige", "navy", "blue", "purple", "gray", "red", "black", "burgundy",
    "coral", "turquoise", "aqua", "gold", "mint"
]
PATTERN_TYPES = [
    "Solid", "Striped", "Floral", "Checked", "Polka Dot",
    "Graphic Print", "Animal Print", "Plaid", "Geometric", "N/A"
]
PATTERN_DESCRIPTIONS = {
    "Floral": ["Tiny floral motifs", "Large floral print"],
    "Striped": ["Thin horizontal stripes", "Thick vertical stripes"],
    "Checked": ["Small check pattern", "Large plaid squares"],
    "Polka Dot": ["Small polka dots", "Large polka dots"],
    "Animal Print": ["Leopard print", "Zebra stripes"],
    "Graphic Print": ["Abstract graphics", "Text slogans"],
    "Geometric": ["Triangle patterns", "Diamond shapes"],
    "Plaid": ["Classic tartan plaid"]
}
SIM_THRESHOLD = 0.2

# === Helper Functions ===
def predict_single(image, options):
    inputs = processor(text=options, images=image, return_tensors="pt", padding=True).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits_per_image.squeeze().cpu()
    probs = logits.softmax(dim=0).numpy()

    if  np.ndim(probs) == 0:  # only one option, scores is scalar
        return options[0], float(probs)
    else:
        idx = int(np.argmax(probs))
        return options[idx], float(probs[idx])


def predict_multi(image, options, threshold=0.20):
    inputs = processor(text=options, images=image, return_tensors="pt", padding=True).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits_per_image.squeeze().cpu()
    probs = logits.softmax(dim=0).numpy()
    results = [opt for opt, score in zip(options, probs) if score >= threshold]
    return results if results else [options[int(np.argmax(probs))]]

def analyze_garment(image_path):
    image = Image.open(image_path).convert("RGB").resize((224, 224))

    clothing_type, _ = predict_single(image, CLOTHING_TYPES)
    gender_suitability = predict_multi(image, GENDER_OPTIONS)
    occasion_suitability = predict_multi(image, OCCASION_OPTIONS)
    sleeve_type, _ = predict_single(image, SLEEVE_TYPES)
    neckline, _ = predict_single(image, NECKLINE_TYPES)
    closure_type, _ = predict_single(image, CLOSURE_TYPES)
    dominant_color, _ = predict_single(image, COLOR_OPTIONS)

    # Secondary color (rerun without dominant)
    color_options = [col for col in COLOR_OPTIONS if col != dominant_color]
    secondary_color, sec_conf = predict_single(image, color_options)
    if sec_conf < SIM_THRESHOLD:
        secondary_color = "N/A"

    pattern_type, _ = predict_single(image, PATTERN_TYPES)
    if pattern_type in PATTERN_DESCRIPTIONS:
        pattern_description, pd_conf = predict_single(image, PATTERN_DESCRIPTIONS[pattern_type])
        if pd_conf < 0.10:
            pattern_description = "N/A"
    else:
        pattern_description = "N/A"

    remarks = ""

    return {
        "clothing_type": clothing_type,
        "gender_suitability": gender_suitability,
        "occasion_suitability": occasion_suitability,
        "sleeve_type": sleeve_type,
        "neckline": neckline,
        "closure_type": closure_type,
        "dominant_color": dominant_color.capitalize(),
        "secondary_color": secondary_color.capitalize(),
        "pattern_type": pattern_type,
        "pattern_description": pattern_description,
        "remarks": remarks
    }

# === Directory Processing ===
SRC_DIR = "/content/viton/clothes_tryon_dataset/test/cloth/"   # Your source image directory
DST_DIR = "features_extracted2"                                 # Output features directory

os.makedirs(DST_DIR, exist_ok=True)
image_exts = (".jpg", ".jpeg", ".png", ".bmp", ".webp")        # Supported formats

for fname in os.listdir(SRC_DIR):
    if fname.lower().endswith(image_exts):
        image_path = os.path.join(SRC_DIR, fname)
        result = analyze_garment(image_path)
        # Save to JSON with same base name
        json_name = os.path.splitext(fname)[0] + ".json"
        json_path = os.path.join(DST_DIR, json_name)
        with open(json_path, "w") as f:
            json.dump(result, f, indent=4)
        print(f"Processed: {fname} -> {json_path}")

print("DONE: All feature JSONs exported to", DST_DIR)
```

```python


from PIL import Image
import torch
import numpy as np

# Load the FashionCLIP model & processor (Hugging Face)
device = "cuda" if torch.cuda.is_available() else "cpu"
model = CLIPModel.from_pretrained("patrickjohncyh/fashion-clip").to(device)
processor = CLIPProcessor.from_pretrained("patrickjohncyh/fashion-clip")

######### DEFINE YOUR LABELS
CLOTHING_TYPES = [
    "Shirt", "T-shirt", "Sweater", "Jacket", "Blazer",
    "Hoodie", "Blouse"
]
GENDER_OPTIONS = [
    "Male", "Female", "Unisex", "Child"
]
OCCASION_OPTIONS = [
    "Casual", "Formal", "Party Wear"
]
SLEEVE_TYPES = [
    "Short Sleeve", "Long Sleeve", "Sleeveless"
]
NECKLINE_TYPES = [
    "Crew Neck", "V-Neck", "Collar", "High Neck", "Hooded", "N/A"
]
CLOSURE_TYPES = [
    "Zipper", "Button", "None", "Tie", "Hook and Eye", "Velcro", "N/A"
]
COLOR_OPTIONS = [
    "brown", "green", "orange", "yellow", "peach", "pink", "teal", "beige", "navy", "blue", "purple", "gray", "red", "black", "burgundy", "coral", "turquoise", "aqua", "gold", "mint"
]
PATTERN_TYPES = [
    "Solid", "Striped", "Floral", "Checked", "Polka Dot",
    "Graphic Print", "Animal Print", "Plaid", "Geometric", "N/A"
]
PATTERN_DESCRIPTIONS = {
    "Floral": ["Tiny floral motifs", "Large floral print"],
    "Striped": ["Thin horizontal stripes", "Thick vertical stripes"],
    "Checked": ["Small check pattern", "Large plaid squares"],
    "Polka Dot": ["Small polka dots", "Large polka dots"],
    "Animal Print": ["Leopard print", "Zebra stripes"],
    "Graphic Print": ["Abstract graphics", "Text slogans"],
    "Geometric": ["Triangle patterns", "Diamond shapes"],
    "Plaid": ["Classic tartan plaid"]
}
SIM_THRESHOLD = 0.2  # Change as desired

###### ATTRIBUTE HEADERS

def predict_single(image, options):
    """Predict the most likely label from options, return label and confidence."""
    inputs = processor(text=options, images=image, return_tensors="pt", padding=True).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits_per_image.squeeze().cpu()
    probs = logits.softmax(dim=0).numpy()
    idx = int(np.argmax(probs))
    return options[idx], float(probs[idx])

def predict_multi(image, options, threshold=0.20):
    """Return all labels above threshold (multi-label), or fallback to best only."""
    inputs = processor(text=options, images=image, return_tensors="pt", padding=True).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits_per_image.squeeze().cpu()
    probs = logits.softmax(dim=0).numpy()
    results = [opt for opt, score in zip(options, probs) if score >= threshold]
    return results if results else [options[int(np.argmax(probs))]]

def analyze_garment(image_path):
    image = Image.open(image_path).convert("RGB").resize((224, 224))

    clothing_type, _ = predict_single(image, CLOTHING_TYPES)
    gender_suitability = predict_multi(image, GENDER_OPTIONS)
    occasion_suitability = predict_multi(image, OCCASION_OPTIONS)
    sleeve_type, _ = predict_single(image, SLEEVE_TYPES)
    neckline, _ = predict_single(image, NECKLINE_TYPES)
    closure_type, _ = predict_single(image, CLOSURE_TYPES)
    dominant_color, _ = predict_single(image, COLOR_OPTIONS)

    # Secondary color: rerun without dominant
    color_options = [col for col in COLOR_OPTIONS if col != dominant_color]
    secondary_color, sec_conf = predict_single(image, color_options)
    if sec_conf < SIM_THRESHOLD:
        secondary_color = "N/A"

    pattern_type, _ = predict_single(image, PATTERN_TYPES)
    # Pattern description if available
    if pattern_type in PATTERN_DESCRIPTIONS:
        pattern_description, pd_conf = predict_single(image, PATTERN_DESCRIPTIONS[pattern_type])
        if pd_conf < 0.10:  # Lower threshold for descriptions
            pattern_description = "N/A"
    else:
        pattern_description = "N/A"

    # (Optional) You can use a separate zero-shot prompt for remarks, e.g. "Describe features...", if you want model-generated text.
    remarks = ""

    return {
        "clothing_type": clothing_type,
        "gender_suitability": gender_suitability,
        "occasion_suitability": occasion_suitability,
        "sleeve_type": sleeve_type,
        "neckline": neckline,
        "closure_type": closure_type,
        "dominant_color": dominant_color.capitalize(),
        "secondary_color": secondary_color.capitalize(),
        "pattern_type": pattern_type,
        "pattern_description": pattern_description,
        "remarks": remarks
    }

# ==== USAGE EXAMPLE ====
# Example: local path or uploaded image in Colab
image_path = "/content/viton/clothes_tryon_dataset/test/cloth/00259_00.jpg"  # CHANGE THIS to your image path

results = analyze_garment(image_path)
import json
print(json.dumps(results, indent=4))

```
